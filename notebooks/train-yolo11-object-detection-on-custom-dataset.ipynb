```python
# ── Cell 3: Train — Optimised for Edge / Android Deployment ─────────────
!yolo task=detect mode=train        \
     model=yolo11n.pt                \
     data='/content/drive/MyDrive/Fish-Freshness-1/data.yaml'   \
     epochs=50                        \
     imgsz=640                        \
     cache=True                       \
     workers=8                        \
     batch=32                         \
     project='/content/drive/MyDrive/Fish-Freshness-1/training_results' \
     name='aqua_scan_v1'               \
     exist_ok=True
```

```python
# ── Cell 4: Validate Trained Model ──────────────────────────────────────
!yolo task=detect mode=val          \
     model='/content/drive/MyDrive/Fish-Freshness-1/training_results/aqua_scan_v1/weights/best.pt' \
     data='/content/drive/MyDrive/Fish-Freshness-1/data.yaml'
```

```python
# ── Cell 5: Export → TFLite int8 (Android / Edge) ───────────────────────
# int8 quantisation gives ~3-4× speed-up on mobile NPU vs float32
!yolo task=detect mode=export       \
     model='/content/drive/MyDrive/Fish-Freshness-1/training_results/aqua_scan_v1/weights/best.pt' \
     format=tflite                    \
     int8=True                        \
     imgsz=640                        \
     data='/content/drive/MyDrive/Fish-Freshness-1/data.yaml'
```

```python
# ── Cell 6: Verify Exported TFLite File ─────────────────────────────────
import os, glob

weights_dir = '/content/drive/MyDrive/Fish-Freshness-1/training_results/aqua_scan_v1/weights'
tflite_files = glob.glob(os.path.join(weights_dir, '**/*.tflite'), recursive=True)

if tflite_files:
    for f in tflite_files:
        size_mb = os.path.getsize(f) / 1e6
        print(f"✅  {f}")
        print(f"    Size: {size_mb:.2f} MB")
else:
    print("❌ No .tflite files found — check export step above.")
```
